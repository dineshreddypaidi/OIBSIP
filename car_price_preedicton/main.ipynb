```python
import pandas as pd
```

```python
df = pd.read_csv('car_data.csv')
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
fuel_uniq = df['Fuel_Type'].unique()
selling_type_uniq = df['Selling_type'].unique()
transmission_uniq = df['Transmission'].unique()
print(f"fuel unique = {fuel_uniq} \nselling_uniq = {selling_type_uniq} \ntransmission uiq = {transmission_uniq}")
```

```python
fuel = {
    'Petrol':0,
    "Diesel":1,
    "CNG":2
}
transmission = {
    "Automatic" : 0,
    "Manual" : 1
}
seller_type = {
    'Individual' : 0,
    'Dealer' : 1
}
```

```python
df['Fuel_Type'] =df['Fuel_Type'].map(fuel)
df['Selling_type'] =df['Selling_type'].map(seller_type)
df['Transmission'] =df['Transmission'].map(transmission)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Car_Name'] = label_encoder.fit_transform(df['Car_Name'])
```

```python
df.head()
```

```python
X = df[['Car_Name','Year', 'Present_Price', 'Driven_kms', 'Fuel_Type', 'Selling_type', 'Transmission', 'Owner']]
Y = df[['Selling_Price']]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=32)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
score = metrics.r2_score(y_test,y_pred)
score
```
